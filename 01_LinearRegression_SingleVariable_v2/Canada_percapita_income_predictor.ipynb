```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
# df
```

```python
df['per capita income (US$)'] #series
```

```python
df[['year']] #dataframe
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')
plt.scatter(df.year,df['per capita income (US$)'],color= 'red',marker = '+')
```

```python
model = LinearRegression()
model.fit(df[['year']],df['per capita income (US$)']) #model.fit(dataframe,series)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[2020]]) #model.predict(dataframe)
```

```python
828.46507522 *2020 + (-1632210.7578554575) #proof
```

```python
model.predict(df[['year']]) #predicting train samples 
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income')

plt.scatter(df.year,df['per capita income (US$)'],color= 'red',marker = '+') #marker: train year vs train PCI 

plt.scatter(df.year,model.predict(df[['year']]),color= 'green',marker = '+') #marker: train year vs predicted PCI
plt.plot(df.year,model.predict(df[['year']])) # predicted regression line(train year vs predicted PCI)
```

```python
df['predicted'] = model.predict(df[['year']]) #adding predicted column to our dataframe for side by side comparison
```
