__Detecting Outliers using quantile method__

```python
min_threshold, max_threshold = df['price'].quantile([0.01,0.999]) # thresholds at 1 % and 99.90 %
min_threshold, max_threshold 
```

```python
df_no_outliers = df[(df['price']>min_threshold) & (df['price']<max_threshold)]
df_no_outliers.shape
```

```python
df_no_outliers['price'].min() # minimum price in dataset
```

```python
max(df_no_outliers['price']) # maximum price in dataset
```

```python
df_no_outliers.describe().price #check for min and max after removing outliers
```
