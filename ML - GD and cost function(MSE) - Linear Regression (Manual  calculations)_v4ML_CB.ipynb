__Understanding the gradient descent function, cost and learning rate for linear regression with mean squared error__

```python
import numpy as np #referred video4 ml- gradient descent and cost function of codebasics
```

```python
# x = [1,2,3,4,5] 
# y =[5,7,9,11,13]
# This is not accepted in list comprehension as it doesnt accept the subtraction of list and int
# Throw this error : TypeError: unsupported operand type(s) for -: 'list' and 'int'. So take a numpy array instead. 

x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
# n = len (x)
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 10000
    n = len (x)
    learning_rate = 0.001
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr # for each value of x, predictions of y are observed. y_pred returns an array of length corresponding to x.
        cost = (1/n) * sum ([val**2 for val in (y - y_predicted)]) # ([5,7,9,11,13] - [0 0 0 0 0])^2 for first iteration
        md = -(2/n)* sum (x*(y - y_predicted)) # partial derivation of error wrt m
        bd = -(2/n)* sum (y - y_predicted) # partial derivation of error wrt b
        m_curr = m_curr - learning_rate*md # 
        b_curr = b_curr - learning_rate*bd
        print (f"y_pred {y_predicted}, m{m_curr}, b{b_curr}, cost {cost}, iteration {i}")
        
```

```python
gradient_descent(x,y)
```

__Observing the converging of gradient descent to minimise the error.__
    
    Increasing the learning rate skips the minima AND COST STARTS INCREASING IN THE OTHER DIRECTION.
    Optimum learning rate is desired. Here reducing the iterations to 10 for better understanding.

```python
def gradient_descent(x,y): #(Linear regression function for 1 weight)
    m_curr = b_curr = 0
    iterations = 10000
    n = len (x)
#     keep iterations as 10 for experimenting with learning rate.Once optimum learning rate is achieved increase the iterations. 
#     learning_rate = 0.10 # cost increasing
#     learning_rate = 0.081 # cost reduces and increases again skipping the minima at iteration 164-165 for 10000 iter. 
    learning_rate = 0.08 # cost started to decrease . So this can be concluded as our optimum learning rate. 
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr # for each value of x predictions of y are observed. y_pred returns an array of length corresponding to x.
        cost = (1/n) * sum ([val**2 for val in (y - y_predicted)]) # ([5,7,9,11,13] - [0 0 0 0 0])^2 for first iteration
        md = -(2/n)* sum (x*(y - y_predicted)) # partial derivation of error wrt m
        bd = -(2/n)* sum (y - y_predicted) # partial derivation of error wrt b
        m_curr = m_curr - learning_rate*md # 
        b_curr = b_curr - learning_rate*bd
        print (f"m {m_curr}, b {b_curr}, cost {cost}, iteration {i}")
```

__Observe that at learning rate = 0.08(in this case : optimum) we approached very close to m = 2 and b = 3 in the last iteration(10000th iteration).__

```python
gradient_descent(x,y)
```
