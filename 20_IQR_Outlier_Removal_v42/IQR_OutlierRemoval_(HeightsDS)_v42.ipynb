```python
import pandas as pd
import numpy as np
import seaborn as sn
from matplotlib import pyplot as plt
%matplotlib inline
df = pd.read_csv('heights.csv')
df
```

```python
df.describe() #25th% 50th% 75th% can be seen in the image above as well
```

__IQR Method__

```python
Q1, Q3 = df['height'].quantile([0.25,0.75]) # thresholds at 0.1 % and 99.90 %
Q1, Q3
#Q1 is the 25th percentile 
#Q3 is the 75th percentile 
# We can also see the same in describe above.
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

__Viewing Outliers__

```python
df[(df.height<lower_limit)|(df.height>upper_limit)]
```

__Creating no outliers dataframe__

```python
df_no_outlier = df[(df.height>lower_limit)&(df.height<upper_limit)]
df_no_outlier
```

```python
df.shape[0]- df_no_outlier.shape[0] #5 outliers removed
```
