## METHOD 1- ONE HOT ENCODING (Using pandas dummy variable)

```python
import pandas as pd
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies = pd.get_dummies(df.town) #using pandas get dummies function to get hot encoded dataframe
dummies
```

```python
merged = pd.concat([df,dummies],axis = 'columns') #joinng two dataframes
merged
```

```python
final = merged.drop(['town','west windsor'],axis = 'columns') 
#dropping town(it is in word format)
#dropping west windsor (to prevent dummy variable trap)
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
X= final.drop('price',axis = 'columns') #creating x as a dataframe for fitting in model (dropping y)
X 
```

```python
y = final.price
y
```

```python
model.fit(X,y) #fit accepts (dataframe/two dimensioanl array, pandas series/list)
```

```python
model.predict([[2800,0,1]]) #predict accepts dataframe/two dimensional array
```

```python
model.score(X,y) #accuracy is predicted
```

## Method 2- One hot Encoding (using sklearns labelEncoder)

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() #creating object
```

```python
dfle = df 
le.fit_transform(dfle.town) #1d array is created from town names.
```

```python
dfle #checking dfle
```

```python
dfle.town = le.fit_transform(dfle.town) #assigning encoded values to town colunm od dfle
dfle
```

```python
X= dfle[['town','area']].values 
type(X)
```

```python
X
```

```python
y = dfle.price
y
```

```python
type(y)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X=X[:,1:]
X
```

```python
model.fit(X,y)
```

```python
pred=model.predict(X)
pred
```

```python
dfle['predictions']= pred
dfle
```

```python
model.score(X,y)
```
