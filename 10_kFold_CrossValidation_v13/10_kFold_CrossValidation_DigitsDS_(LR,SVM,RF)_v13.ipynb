```python
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.datasets import load_digits
import pandas as pd
import numpy as np
```

```python
digits = load_digits()
dir(digits)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target) # (flattened 2d array,1d array)
```

__Logistic Regression__

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

__Support Vector Machines__

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
```

```python
svm.score(X_test,y_test)
```

__Randomforest__ 

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50) #default no of trees is 100 ('n_estimators': 100). Observe next cell o/p
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

__Train test split based on indices with kfold technique and finding the scores for each model__

```python
scores_lr = []
scores_svm = []
scores_rf= []
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits= 3) # default n_splits = 5
kf
```

```python
#trial
for i,j in kf.split(digits.data): #Iterates for n_splits= 3 ad returns the indices.Refer next cell for implementation.
    print (i.shape) #returns indices for train set
    print (j.shape) #returns indices for test set
```

```python
for i,j in kf.split(digits.data): #Iterates for n_splits= 3 ad returns the indices.Refer next cell for implementation.
    X_train, X_test, y_train, y_test= digits.data[i],digits.data[j],digits.target[j],digits.target[j]
    scores_lr.append(lr.score(X_test,y_test))
    scores_svm.append(svm.score(X_test,y_test))
    scores_rf.append(rf.score(X_test,y_test))
```
